## Code to scrape customer reviews from Amazon

```python
from selenium import webdriver

# Set driver as Chrome Driver for interaction with webpage
driver = webdriver.Chrome('C:\\Users\\AMAN\\Desktop\\machine_learning\\case study\\73 string\\chromedriver_win32\\chromedriver.exe')

#Initial navigation 
driver.get("https://www.amazon.in/gp/most-gifted/electronics/1805560031/ref=zg_mg_pg_2?ie=UTF8&pg=1")
```

#### Scrape all the mobile phone urls and mobile name from the page

```python
allmobileLink = []
allmobilename = []

#Navigate to all the pages present

pagination = 'true'
while pagination == 'true':
    
    # Scrape all the urls of mobile phone present
    mobileLinksList = driver.find_elements_by_xpath("//*[@id='zg-ordered-list']/li/span/div/span/a[@class='a-link-normal']")
    
    #Scrape respective names of mobile phones
    mobileNameList = driver.find_elements_by_xpath("//*[@id='zg-ordered-list']/li/span/div/span/a[@class='a-link-normal']/div")
    
    for i in mobileLinksList:
        mobile = i.get_attribute('href')    
        allmobileLink.append(mobile)
    for i in mobileNameList:
        mobile = i.text    
        allmobilename.append(mobile)
    try:
        #If pagination is there click on next page
        driver.find_element_by_xpath("//*[@id='zg-center-div']/div[2]/div/ul/li/a[text()='Next page']").click()
        while status!='complete':
            status = driver.execute_script('return document.readyState')
    except:
        break
        # Pagination ends
```

```python
# Regex to obtain mobile name from full title
import re
allmobilename = [re.sub(r'-.*|\|.*|\s‚Äì.*','',name) for name in allmobilename]
allmobilename = [re.sub(r'\s$','',name) for name in allmobilename]
```

#### Navigate to each mobile link and scrape the reviews

```python
mobile_review = {}

# Wait until page loads completely
def load_status():
    status = ''
    while status!='complete':
        status = driver.execute_script('return document.readyState')

for i in range(len(allmobileLink)):
    review = []
    driver.get(allmobileLink[i])    
    load_status()
        
    try:
        # Click on see all reviews to display top 10 reviews
        driver.find_element_by_xpath("//a[text()='See all reviews']").click()
        load_status()
        reviewList = driver.find_elements_by_xpath('//*[contains(@id,"customer_review")]/div/a/span')
    except:
        #If less reviews are there and See all reviews link is not present
        reviewList = driver.find_elements_by_xpath('//*[contains(@id,"customer_review")]/div/a/span')   
    
    
    
    for j in reviewList:
        review.append(j.text)
    
    #Dictionary having key as mobile name and value as list od=f top 10 reviews
    mobile_review[allmobilename[i]]=review
```

### create data frame and export as csv all the reviews scraped

```python
#Convert Dictionary to data Frame

import pandas as pd
df_mobile_review = pd.DataFrame.from_dict(mobile_review,orient ='index')
```

```python
#Export as CSV file

df_mobile_review.to_csv('mobile_review.csv')
```

CSV Details:

Index is Mobile Name

Columns are Top 10 reviews

Null rows represent that reviews are not present in website
