# Data visualization of Sentiment Analysis Result

### Plot representing count of each Sentiment type present 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Review_Sentiment_Score.csv')
p = sns.countplot(data=df, x = 'Sentiment')
```

### Plot representing mobile name and corresponding Sentiment value

```python
sns.set(rc={'figure.figsize':(4.7,40.27)})
p = sns.barplot(data=df,y='Mobile',x='Sentiment')
```

### Distribution Plot showing distribution of Sentiments which follows Gaussian distribution

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(df['Sentiment'])
```
